````python
# df1 = 2014
# df2 = fmed
# Leemos los csv existentes
df1_dict, df1_data, df_instructions, df2_data = new.read_data()
# Creamos un nuevo dataframe vacio para guardar los datos recodificados
df2_data_new = pd.DataFrame()
# Creamos un nuevo dataframe con la estructura diccionario de datos correspondiente a los datos recodificados
df2_dict_new = new.create_dict()
num_rows = len(df_instructions)
print(num_rows)
# Guardamos los indices y nombres de las columnas que no se pudieran procesar.
no_encontrado = {'name_column': [], 'id_column': []}
for n in range(num_rows):
    print(f"\n----------------------------\nESTAMOS EN EL RENGLON DE INSTRUCCIONES: {n} - {df_instructions['campo_unificado'].iloc[n]}")
    new_column_name, description_df1, description_df2, description_final, value_options, options_updated, actions, df1_column_name, df2_column_name, category = new.read_instructions(df_instructions, n)
    value_options, actions = new.convert_to_dict(value_options, actions)
    if df2_column_name in df2_data.columns:
        # Perform the actions on the data and get the new dictionary
        code, description, options, category = new.apply_actions(df2_data, df2_data_new, new_column_name, description_df1, description_df2, description_final, value_options, options_updated, actions, df1_column_name, df2_column_name, category)
        df2_dict_new = new.add_to_new_dict(df2_dict_new, code, description, options, category)
    else:
        print(f"La variable {df2_column_name} en el renglon {n} no se encuentra en el dataframe")
        no_encontrado['name_column'].append(df2_column_name)
        no_encontrado['id_column'].append(n)   
````

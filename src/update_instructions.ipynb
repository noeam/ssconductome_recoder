```python
import pandas as pd
```

```python
df_instructions = pd.read_csv("~/ServicioSocial/ssocial/data/instruccionesH3.csv")
df_instructions
```

```python
dict_fmed = pd.read_csv("~/ServicioSocial/ssocial/data/diccionarios/fmed_diccionario.csv")
dict_fmed.head(10)
```

```python
df_instructions_new = pd.DataFrame()
df_instructions_new['campo_unificado'] = df_instructions['campo_unificado']
df_instructions_new['description_2014'] = df_instructions['description_2014']
df_instructions_new['acciones'] = df_instructions['acciones']
df_instructions_new['options'] = df_instructions['options']
df_instructions_new['code_2014'] = df_instructions['code_2014']
df_instructions_new['code_fmed'] = df_instructions['code_fmed']
df_instructions_new
```

```python
# TEST para identificar la descripcion correspondiente a la variable 'a9'
item = dict_fmed.loc[dict_fmed["campo_original"] == "a9", "question"].item() #iloc[0]
print(type(item))
print(item)
```

```python
df_instructions_new['description_fmed'] = df_instructions['description_2014']
df_instructions_new['code_fmed_completo'] = df_instructions['campo_unificado']
df_instructions_new['subcategoria'] = df_instructions['campo_unificado']
for index, row in df_instructions_new.iterrows():
    #print(index, row["code_fmed"])
    description_fmed = dict_fmed.loc[dict_fmed["campo_original"] == row["code_fmed"], "question"].iloc[0]
    df_instructions_new['description_fmed'][index] = description_fmed
    new_code = dict_fmed.loc[dict_fmed["campo_original"] == row["code_fmed"], "campo_unificado"].iloc[0]
    df_instructions_new['code_fmed_completo'][index] = new_code
    subcategory = dict_fmed.loc[dict_fmed["campo_original"] == row["code_fmed"], "seccion_instrumento"].iloc[0]
    df_instructions_new['subcategoria'][index] = subcategory
    
df_instructions_new
```

```python
df_instructions_new.to_csv("~/ServicioSocial/ssocial/data/new_instruccionesH3.csv")
```
